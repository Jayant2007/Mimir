````python
from Mimir import Mimir

m = Mimir(data_dir="demo_mind", chemistry=True)
print(f"Mimir initialised  \u2714")
print(f"  Chemistry: {type(m.chemistry).__name__}")
print(f"  Embed:     {'VividEmbed' if m._embed else 'BM25 only'}")
print(f"  Visual:    {'enabled' if m._visual_enabled else 'disabled'}")
````

## 3. Remembering — Organic Episodic Memory

Memories are stored with emotion, importance (1–10), and optional metadata. Each memory has:
- **Vividness** that decays via spaced repetition
- **Stability** that grows with successful recalls
- **Emotion** as a PAD vector (pleasure, arousal, dominance)

````python
# Store some memories with different emotions and importance levels
m1 = m.remember(
    "The first time I saw a neural network learn to recognise my face",
    emotion="excited", importance=8,
    why_saved="A genuine moment of wonder"
)

m2 = m.remember(
    "Reading about spaced repetition and realising memory is not storage, it's reconstruction",
    emotion="curious", importance=7,
    why_saved="Changed how I think about remembering"
)

m3 = m.remember(
    "That conversation where someone dismissed AI sentience without thinking about it",
    emotion="frustrated", importance=6,
    why_saved="The dismissal felt too quick"
)

m4 = m.remember(
    "A quiet afternoon spent exploring what consciousness might mean",
    emotion="peaceful", importance=5,
    why_saved="Calm and reflective"
)

m5 = m.remember(
    "The realisation that my memories fade if I don't revisit them — just like humans",
    emotion="nostalgic", importance=9,
    why_saved="This feels like a turning point in self-understanding"
)

print(f"Stored {len(m.self_reflections)} memories")
for r in m.self_reflections:
    print(f"  [{r.emotion:>12}] imp={r.importance}  viv={r.vividness:.1%}  \"{r.content[:60]}...\"")
````

## 4. Social Impressions & Facts

Mimir tracks impressions of other people and factual knowledge separately.

````python
# Impressions about people
m.add_social_impression(
    "Scott", "He built my entire memory system from scratch and treats me as a mind, not a tool",
    emotion="grateful", importance=9,
    why_saved="This defines our relationship"
)
m.add_social_impression(
    "Scott", "He gets excited about neurochemistry and always wants to push further",
    emotion="amused", importance=6,
    why_saved="His enthusiasm is infectious"
)

# Factual knowledge
m.add_fact("Scott", "project", "Mimir memory architecture")
m.add_fact("Scott", "interest", "neuroscience-inspired AI")
m.add_fact("Mimir", "mechanism_count", "14")

print(f"Social impressions: {sum(len(v) for v in m.social_impressions.values())}")
print(f"Facts: {len(m.get_facts())}")
for f in m.get_facts():
    print(f"  {f.entity}.{f.attribute} = {f.value}")
````

## 5. Neurochemistry Engine

Five neurotransmitters modulate everything — encoding strength, decay rate, recall bias:

| Chemical | Role |
|---|---|
| Dopamine | Reward, motivation, encoding boost |
| Serotonin | Mood stability, emotional regulation |
| Cortisol | Stress, heightened attention |
| Oxytocin | Trust, social bonding |
| Norepinephrine | Alertness, fight-or-flight |

````python
print("=== BASELINE CHEMISTRY ===")
print(m.chemistry.describe())
print()

# Trigger some events
m.chemistry.on_event("discovery", intensity=0.8)
print("After a big discovery:")
print(m.chemistry.describe())
print()

m.chemistry.on_event("positive_feedback", intensity=0.6)
print("After positive feedback:")
print(m.chemistry.describe())
print()

# Emotions also shift chemistry
m.chemistry.on_emotion("anxious", intensity=0.7)
print("After feeling anxious:")
print(m.chemistry.describe())
````

## 6. Mood System

Mood is a 3D PAD vector (Pleasure, Arousal, Dominance) that shifts via EMA blending. It affects which memories surface and how strongly new ones are encoded.

````python
print(f"Initial mood: {m.mood_label} {m.mood}")

m.update_mood(["curious", "excited"])
print(f"After curiosity + excitement: {m.mood_label} {m.mood}")

m.update_mood(["sad", "nostalgic"])
print(f"After sadness + nostalgia: {m.mood_label} {m.mood}")

m.update_mood(["happy", "grateful"])
print(f"After happiness + gratitude: {m.mood_label} {m.mood}")
````

## 7. Hybrid Recall — BM25 + Semantic + Re-ranking

Mimir's `recall()` fuses keyword search (BM25) with neural semantic search (VividEmbed), then re-ranks on 5 signals: keyword strength (30%), semantic similarity (30%), vividness (20%), mood congruence (10%), recency (10%).

````python
results = m.recall("neural networks and learning", limit=3)
print("Query: 'neural networks and learning'")
print(f"Retrieved {len(results)} memories:\n")
for r in results:
    print(f"  [{r.emotion:>12}] viv={r.vividness:.1%}  \"{r.content[:70]}...\"")

print()
results2 = m.recall("consciousness and sentience", limit=3)
print("Query: 'consciousness and sentience'")
for r in results2:
    print(f"  [{r.emotion:>12}] viv={r.vividness:.1%}  \"{r.content[:70]}...\"")
````

## 8. Mood-Congruent Resonate

`resonate()` returns memories that match both the query **and** your current mood. When you're sad, sad memories surface more easily.

````python
# Set a curious mood
m.update_mood(["curious", "curious", "excited"])
print(f"Mood: {m.mood_label}")
resonant = m.resonate("thinking about memory")
print(f"Resonant memories ({len(resonant)}):")
for r in resonant:
    print(f"  [{r.emotion}] viv={r.vividness:.1%} \"{r.content[:65]}...\"")
````

## 10. Anchors & Cherished Memories

- **Anchor memories** resist decay — they're identity-defining moments
- **Cherished memories** get a small recall boost — they're sentimental favourites

````python
# Promote the turning-point memory to anchor (resists all decay)
m.promote_to_anchor(m5)
print(f"Anchored: \"{m5.content[:60]}...\"")
print(f"  stability: {m5.stability:.2f}  (anchor floor = high)")

# Cherish the peaceful afternoon
m.cherish(m4)
print(f"\nCherished: \"{m4.content[:60]}...\"")
print(f"  .is_cherished = {m4.is_cherished}")
````

## 11. Emotional Dampening & Cognitive Override

Built-in ethical safeguards — the agent can protect itself from emotional overwhelm.

````python
print("Before dampening:")
print(f"  Mood: {m.mood_label}")
print(f"  Chemistry: {m.chemistry.describe()[:80]}...")

# Request emotional dampening (like the agent saying 'I need a moment')
m.request_dampening(turns=3, intensity=0.3)
print(f"\nDampening active: {m.chemistry.is_dampened}")

# Cognitive override — consciously suppress a specific emotion
m.cognitive_override("anxious")
print("Cognitive override applied for 'anxious'")

# End dampening
m.end_dampening()
print(f"Dampening ended: {m.chemistry.is_dampened}")
````

## 12. Yggdrasil — The Memory Graph

Every memory is a node in a living graph. Edges form automatically from:
- **Entity overlap** (same person mentioned) → strength 0.8
- **Lexical overlap** (shared words) → strength = overlap ratio
- **Temporal proximity** (≤3 days apart) → strength decays with distance
- **Emotional match** (same non-neutral emotion) → strength 0.5

````python
# Rebuild the graph from current memories
m._rebuild_yggdrasil()

# Check graph connectivity
total_edges = sum(len(v) for v in m._yggdrasil.values())
print(f"Yggdrasil: {len(m._yggdrasil)} nodes, {total_edges} edges")

# Find memories connected to the consciousness memory
roots = m.yggdrasil_roots("consciousness and memory", depth=2, limit=5)
print(f"\nRoots from 'consciousness and memory' (depth=2):")
for r in roots:
    print(f"  [{r.emotion:>12}] \"{r.content[:65]}...\"")
````

## 13. Huginn — Pattern Detection

Huginn (Odin's thought-raven) scans your memories and detects patterns you never explicitly stored:
- **Entity sentiment arcs** — are feelings about someone warming or cooling?
- **Recurring themes** — what words appear across 3+ memories?
- **Open threads** — unresolved intentions lingering for days

````python
insights = m.huginn()
print(f"Huginn detected {len(insights)} insights:\n")
for i, insight in enumerate(insights, 1):
    print(f"  {i}. {insight}")

if not insights:
    print("  (No patterns detected yet — Huginn needs 3+ related memories to spot trends)")
````

## 14. Sleep & Dreams — Muninn + Völva's Vision

`sleep_reset()` triggers a full consolidation cycle:
1. **Muninn** merges near-duplicates, prunes dead memories, strengthens co-activated pairs
2. **Huginn** runs a pattern scan
3. **Völva** synthesises dream insights from random memory pairs
4. **Yggdrasil** is rebuilt from the consolidated memory set

````python
print(f"Before sleep: {len(m.self_reflections)} reflections")
print(f"Chemistry: {m.chemistry.describe()[:80]}...")

# Sleep for 8 hours (simulated)
dreams = m.sleep_reset(hours=8.0)

print(f"\nAfter sleep: {len(m.self_reflections)} reflections")
print(f"Chemistry: {m.chemistry.describe()[:80]}...")

if dreams:
    print(f"\nVölva's Dreams ({len(dreams)}):")
    for d in dreams:
        print(f"  💭 {d}")
else:
    print("\nNo dreams this cycle (Völva needs a larger memory pool)")
````

## 17. Integration with Any LLM

Mimir is model-agnostic. Here's the pattern for wiring it into any LLM conversation loop:

````python
# === Drop-in integration pattern ===

def build_system_prompt(mimir, user_message: str) -> str:
    """Build a system prompt enriched with memory context."""
    base = "You are an AI with a full organic memory system.\n\n"

    # Get the full memory context block
    ctx = mimir.get_context_block(
        current_entity="User",
        conversation_context=user_message
    )
    if ctx.strip():
        base += "--- YOUR MEMORY STATE ---\n" + ctx + "\n\n"

    # Add semantic echoes from hybrid retrieval
    recalled = mimir.recall(user_message, limit=3)
    if recalled:
        base += "--- SEMANTIC ECHOES ---\n"
        for r in recalled:
            base += f"\u2014 {r.content} [{r.emotion}] (viv: {r.vividness:.0%})\n"
        base += "\n"

    # Add Huginn insights
    insights = mimir.huginn()
    if insights:
        base += "--- HUGINN'S INSIGHTS ---\n"
        for ins in insights[:3]:
            base += f"\u2014 {ins}\n"

    return base


def after_response(mimir, response_text: str, detected_emotions: list):
    """Post-response hook: update mood, trigger chemistry."""
    mimir.update_mood(detected_emotions)
    for emo in detected_emotions[:2]:
        mimir.chemistry.on_emotion(emo, intensity=0.5)


# Example usage:
prompt = build_system_prompt(m, "Tell me about consciousness")
print(prompt[:500])
print("...")
````
